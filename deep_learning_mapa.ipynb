# A coupled map lattice
The characterization of space-time chaos is one of the most important problems in the nonlinear science of our time. It is essential not only for understanding the nature of fluid turbulence, optics, physics and solid-state chemistry, but also for a study of coupled dynamic systems in biology and technology. There is a hierarchy of spatially extended dynamical systems which can be used to investigate spatio-temporal dynamical features, from cellular automata, where space, time and state variable are all discrete, to partial differential equations, where the same variables are continuous. One outstanding type of spatio-temporal system is a coupled map lattice(CML), where the space and time are discrete, but with a continuous state variable.

Its relative simplicity turns numerical computations easier and faster to perform using personal computers. They have been extensively used to illustrate many interesting features like chimeras.

A CML is a dynamic system with discrete time and space and a continuous state. Although there are several possible models, we restrict ourselves to the following diffusive coupling model here.

\begin{eqnarray} 
u^{(i)}_{n+1} = (1 - \varepsilon)f(u^{(i)}_{n})+\frac{\varepsilon}{2}\lbrace f(u^{(i+1)}_{n})+f(u^{(i-1)}_{n})\rbrace ,
\end{eqnarray}  

where $u^{(i)}_{n}$ is the discrete-time state variable $n=0,1,2,...$ and $i=1,2,...,N$ is a discrete position in a one-dimensional lattice of size $ N $.\\
\indent The temporal dynamics in each spatial position is described by a quadratic map
\beq
f(u) = (1+\alpha)u -\beta u^{2}, \qquad (0\leq u \leq 1)
\eeq
where $\alpha$ and $\beta$ are a system parameter and $ \varepsilon > 0 $ is the strength of coupling. Next, we keep $ \beta = 0.1 $ constant and use $ \alpha $ as an adjustable parameter. The general characteristics of the space-time dynamics remain practically unchanged for the values of $ 0 <\beta <1 $. We will consider random initial conditions in the range $ 0 \leq u^{(i)}_{0} \leq 1.0$ and periodic boundary conditions: $ u^{(i)}_{n} = u^{(N+1)}_{n}$.

 As has been reported, the model shows a rich variety of phenomena, such as period-doubling bifurcations of kinks, spatial bifurcation, spatio-temporal intermittency, and pattern competition intermittency. The spatio-temporal dynamics can be seen in Fig.1, which summarizes the main patterns as a function of the parameters ($ \alpha$ and $ \varepsilon $). In Fig.(a) we have the periodic pattern, consisting of several regions with different periods of oscillation and different spatial lengths. In Fig. 1(b), we have the formation of the frozen random pattern that has a mixture of periodic domains with chaotic domains and these domains are fixed after the transient. Fig. 1(c) is characterized mainly by short period attractors (less than 10) or a predominance of ordered patterns after the transient. Some regions exhibit dominant patterns of well-defined periodicity, such as (2,4,8) (pattern selection). Chaotic defects (Fig. 1(d)) are chaotic regions located in a zigzag pattern, where these chaotic regions move through the lattice. These chaotic defects are transient, that is, after a long time, the pattern has only a zigzag dynamic, we see they disappear when they collide with each other. Space-time intermittency 1(e)) has an intermittency between chaotic regions and a zigzag pattern . Already fully developed space-time chaos, there are only chaotic regions in the entire lattice both in time and space(Fig. 1(f)). In Fig. 2 it is shown which values of $ \alpha $ with $ (\varepsilon = \beta = 0.1 $) the formation of these patterns occurs.

# Quantifiers

The first quantifier used to characterize each of these patterns is the local order parameter $ R_{i}$. It is a quantifier that determines the degree of coherence of the lattice. We introduce a phase for the jth map from the following definition:
\begin{eqnarray}  
\sin \psi_{j} = \dfrac{2x^{j} -\max_{j}\lbrace x^{j}\rbrace -\min_{j}\lbrace x^{j}\rbrace}{\max_{j}\lbrace x^{j}\rbrace -\min_{j}\lbrace x^{j}\rbrace}, 
\end{eqnarray} 
for $(j=1,...,N)$. The values $\max_{j}\lbrace x^{j}\rbrace$ and $\min_{j}\lbrace x^{j}\rbrace$ are, respectively, the maximum and minimum values of the state variable for a network (snapshot), so that a spatial half-cycle is mapped in the phase interval $ [- \pi/2,\pi/2] $. The magnitude of the local order parameter is defined as: 
\begin{eqnarray} 
R_{i} = \lim_{N\longrightarrow\infty}\frac{1}{2\delta (N)}\left\vert \sum_{j\in C}e^{i\psi_{j}}\right\vert, 
\end{eqnarray} 
for $(i=1,...,N)$. The sum is restricted to the range of j-values, so that, 
\begin{eqnarray}  
C:\left\vert \frac{j}{N} - \frac{i}{N}\right\vert \leq \delta (N), 
\end{eqnarray}  
where $\delta (N)\longrightarrow 0$ when $N\longrightarrow\infty$.\\
 For the $R_{i}\approx 1.0$ we have sites belonging to coherent domains. When $R_{i}$ takes values less than $1.0$, we have sites in inconsistent domains.
 
We can define a new parameter known as the $ P $ degree of coherence, which determines the number of sites with $ R_{i} \approx 1.0 $ in relation to the size of the lattice, that is, we add all sites with $ R_{i} \approx 1.0 $ and divided by the size of the network (snapshot) Fig. 2(a). 
 Another quantifier widely used to determine the dynamics of a lattice of coupled maps is the use of the Lyapunov spectrum $\lbrace \lambda_{i}\rbrace^{N}_{i=1}$, where $\lambda_{i}>0$ we have a chaotic dynamic. The normalized sum of the positive Lyapunov exponents are a numerical estimate for the Kolmogorov-Sinai density (KS) Fig.2(b). We can also work with the maximum exponent of Lyapunov ${\lambda_{\max}}$ as shown in Fig.2(c). In Fig. 2(e), we define the difference delta function ($ \Delta $) that determines the transitorily of the chaotic dynamics, that is, it measures how much the chaotic dynamics decreases with time. For example, when we work with chaotic defects, we know that they are transient, that is, with time they disappear from the lattice, as they are chaotic, it is expected that the entropy KS will decrease. The difference delta calculation is simply
 
\begin{eqnarray} 
\Delta =\dfrac{KS_{i} - KS_{f}}{KS_{i}+KS_{f}}, 
\end{eqnarray} 

where $KS_{i}$ is the measure of the initial $ KS $ and $ KS_{f} $ is the measure of the final $ KS $. In 2(e) we use $ n = 10\,000 $ and final $ n = 100\, 000 $ as the initial time. 

The Shibata exponent ($\zeta$) (Fig. 2(d)) as a dynamic quantifier. It determines the degree of disorder of the variable $u$ in a discrete time $k$ and can be defined as:
\begin{eqnarray} 
\zeta_{k} = \frac{1}{N}\ln\vert \det B^{k,N}\vert ,
\end{eqnarray}  
where $B^{k,N}$ is the Jacobian matrix of the lattice
\begin{eqnarray} 
B^{k,N}_{i,j} = \frac{\partial u^{k+1}_{i}}{\partial u^{k}_{j}} \qquad (i,j=1,2, \ldots), 
\end{eqnarray} 
where $k$ and $j$ are time and discrete space, respectively. However, as the dynamic system evolves over time, the degree of disorder represented by $ \lambda_ {k} $ varies over time, so it is best to work with the time average of the average Shibata exponent 
\begin{eqnarray} 
\Lambda = \frac{1}{n_{k}}\sum^{n_{k}-1}_{k=1} \zeta_{k}, 
\end{eqnarray} 
where $ n_{k} $ is the time interval used to average.

It should be emphasized that these average Shibata exponent are different of Lyapunov expoente, calculated in the phase space, and that they measure the exponential rate of separation between two initially close trajectories. As defined by Shibata, the average exponent must be taken as a local measure of disorder and does not correspond to exponential separation rates. However, the general interpretation is basically the same: the higher the values of $\Lambda$, the greater the degree of disorder in the latticer.

# Machine Learning

ML is the field of study that gives the computer the ability to learn without being previously programmed. Depending on the problems that machine learning can face, it is useful to define three major classes: supervised learning, unsupervised learning and reinforcement learning
. In this article, we will focus only on supervised learning.
•
Supervised learning is a subcategory of ML that focuses on learning a classification or regression model from known training data,  where we know a priori the specific characteristics of each class, respectively. The principle of this algorithm is to find a “hypothesis” function $h$ that approximates the unknown function $f$, using a classified training data set for this purpose. From this learned function, being able to predict the class of new unknown characteristics.
•
\begin{eqnarray} 
h :\mathcal{X} \longrightarrow \mathcal{Y},
\end{eqnarray}
•
where $\mathcal{X} = \mathbb{R}^{m}$ and $\mathcal{Y} = (1, ..., k)$ is a vector of integers with class labels $k$.  
In regression, the task of learning a function
•
\begin{eqnarray} 
•
\mathbb{D}:\mathbb{R}^{m} \longrightarrow R
\end{eqnarray}
given the training set
\begin{eqnarray} 
\mathbb{D}:{\langle x^{[i]},y^{[i]} \rangle ,i=1,...,n},
\end{eqnarray}
where we denote the $\textit{i}$th  training example as $\langle x^{[i]},y^{[i]}\rangle$.\\
\indent We can represent the training set in the form of a matrix $\textbf{X} \in \mathbb{R}^{n \times m} $, where each row represents a training sample and each column, the number of characteristics of that sample.
•
\begin{eqnarray}
\textbf{X}=
\begin{Bmatrix}
x_{1}^{[1]} &x_{2}^{[1]} & \ldots & x_{m}^{[1]}\\
x_{1}^{[2]} & x_{2}^{[2]}&\ldots & x_{m}^{[2]}\\
\vdots & \vdots & \ddots &\vdots\\
x_{1}^{[n]} & x_{2}^{[n]}&\ldots & x_{m}^{[n]},\\
\end{Bmatrix}   
\end{eqnarray}
•
where stores $n$ training examples and $m$ the feature number.
•
Note that to distinguish the characteristic index and the training example index, we will use a superscript notation enclosed in square brackets to refer to the $\textit{i}th$  training example and a regular underline notation to refer to the jth characteristic.
The corresponding labels are represented in a column vector $\textbf{y}$, $\textbf{y} \in \mathbb{R}^{m}$
\begin{eqnarray}
\textbf{y}=
\begin{pmatrix}
y^{[1]}\\
y^{[2]} \\
 \vdots\\
y^{[n]}.\\
\end{pmatrix}
\end{eqnarray}
•
 When the values of $ y_ {i} $ are defined by a limited number of discrete values, we have a classification problem.


For this preparation, we separated $70\%$ of these samples for training (2457 samples) and $30\%$ (1053 samples) of the data for a test step. For good sampling preparation, it must be done randomly, that is, this random separation is performed to have a greater uniformity in the data.


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123, shuffle=True)
print(X_train.shape)
print(X_test.shape)
```

```python
y_train
```

## Normalizing the data

```python
mu, sigma = X_train.mean(axis=0), X_train.std(axis=0)
print(mu,sigma)

X_train = (X_train - mu) / sigma
X_test = (X_test - mu) / sigma
```

## Deep learning architecture



```python
import tensorflow as tf
```

### Building MLP


```python
model = tf.keras.models.Sequential() #PARA CRIAR O MODELO SENQUENCIAL
model.add(tf.keras.Input(shape=5, )) #CAMADA DE ENTRADA
model.add(tf.keras.layers.Dense(units=128, activation='relu')) #PRIMEIRA CAMADA ESCONDIDA
#model.add(tf.keras.layers.Dropout(0.5)) #USA DROPOUT PARA DESLIGAR 20 PORCENTO DOS NEURONIOS
model.add(tf.keras.layers.Dense(units=75, activation='relu')) #SEGUNDA CAMADA ESCONDIDA
#model.add(tf.keras.layers.Dropout(0.2)) #USA DROPOUT PARA DESLIGAR 20 PORCENTO DOS NEURONIOS
model.add(tf.keras.layers.Dense(units=7, activation='softmax')) #USA A FUNCAO SOFTMAX NA ULTIMA CAMADA

model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])

model.summary()
```

```python
history = model.fit(X_train, y_train, batch_size=100, validation_split=0.1, epochs=200)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['sparse_categorical_accuracy'])
plt.plot(history.history['val_sparse_categorical_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

### Evaluation

```python
y_pred = model.predict(X_test)
```

```python
test_loss, test_accuracy = model.evaluate(X_test, y_test)

print('\nTest accuracy:', test_accuracy)
```

```python
pred_test = []
for i in range(y_pred.shape[0]):
    pred_test.append(np.argmax(y_pred[i]))
pred_test = np.array(pred_test)
```

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test, pred_test), cmap='OrRd', annot=True, fmt='2.0f')
plt.title('Confusion Matrix')
plt.ylabel('Prediction')
plt.xlabel('True')
plt.show()
```
